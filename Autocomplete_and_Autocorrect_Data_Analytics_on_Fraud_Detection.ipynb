#**NLP Preprocessing**

Since the dataset only has numerical features, we will generate a sample text data based on the numerical features.

```python
import random

# Sample words
sample_words = ['transaction', 'amount', 'card', 'fraud', 'payment', 'purchase', 'user', 'account']

# Generating synthetic sentences
def generate_synthetic_text(n):
    return [' '.join(random.choices(sample_words, k=random.randint(5, 15))) for _ in range(n)]

synthetic_data = pd.DataFrame({
    'text': generate_synthetic_text(len(df))
})

synthetic_data.head()
```

```python
import nltk
from nltk.corpus import stopwords
import re

nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    text = text.lower()
    words = text.split()
    return ' '.join(words)

synthetic_data['cleaned_text'] = synthetic_data['text'].apply(preprocess_text)
```

#**Implementing Autocomplete and Autocorrect using Bigram and Trigram model**

**Bigram model**

```python
from collections import defaultdict

texts = synthetic_data['cleaned_text'].tolist()

bigrams = defaultdict(list)

for text in texts:
    words = text.split()
    for i in range(len(words) - 1):
        bigrams[words[i]].append(words[i + 1])

def bigram_autocomplete(prefix, num_predictions=3):
    next_words = bigrams.get(prefix, [])

    next_word_freq = pd.Series(next_words).value_counts().head(num_predictions)
    return next_word_freq.index.tolist()

print(bigram_autocomplete("transaction"))
```

```python
# Testing with different prefixes
print(bigram_autocomplete("transaction"))
print(bigram_autocomplete("user"))
print(bigram_autocomplete("amount"))
```

**Metrics analysis**

```python
true_labels = ["card", "payment", "fraud"]
predicted_labels = ["card", "amount", "transaction"]
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

y_true = [1 if true in predicted_labels else 0 for true in true_labels]
y_pred = [1 if pred in true_labels else 0 for pred in predicted_labels]

# metrics
accuracy = accuracy_score(y_true, y_pred)
precision = precision_score(y_true, y_pred)
recall = recall_score(y_true, y_pred)
f1 = f1_score(y_true, y_pred)
print(f"Accuracy: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1 Score: {f1:.2f}")
```

**Trigram Model**

```python
from collections import defaultdict

texts = synthetic_data['cleaned_text'].tolist()

trigrams = defaultdict(list)

for text in texts:
    words = text.split()
    for i in range(len(words) - 2):
        sequence = (words[i], words[i + 1])
        next_word = words[i + 2]
        trigrams[sequence].append(next_word)

def trigram_autocomplete(prefix, num_predictions=3):
    prefix_words = tuple(prefix.split()[-2:])
    next_words = trigrams.get(prefix_words, [])

    next_word_freq = pd.Series(next_words).value_counts().head(num_predictions)
    return next_word_freq.index.tolist()

print(trigram_autocomplete("transaction card"))
```

```python
print(trigram_autocomplete("transaction card"))
print(trigram_autocomplete("user payment"))
print(trigram_autocomplete("amount card"))
```

#**Autocorrect**

```python
import difflib

# Sample words
dictionary = ["transaction", "transfer", "card", "amount", "payment", "user", "fraud", "report", "details","account","valid"]

def autocorrect(word, dictionary, num_suggestions=3):

    suggestions = difflib.get_close_matches(word, dictionary, n=num_suggestions, cutoff=0.6)
    return suggestions

print(autocorrect("transction", dictionary))  # Misspelled word
print(autocorrect("frad", dictionary))
```

```python
from collections import Counter
#sort by frequency
word_frequency = Counter()
for text in texts:
    word_frequency.update(text.split())

dictionary = list(word_frequency.keys())
```

```python
def autocorrect_with_frequency(word, dictionary, word_frequency, num_suggestions=3):
    suggestions = difflib.get_close_matches(word, dictionary, n=10, cutoff=0.6)

    suggestions.sort(key=lambda x: word_frequency[x], reverse=True)

    return suggestions[:num_suggestions]

print(autocorrect_with_frequency("transction", dictionary, word_frequency))
```
